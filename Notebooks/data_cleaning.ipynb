```python
# Import dependencies
import os
import pandas as pd
```

```python
# Import dataframe and add headers
full_df = pd.read_csv("../Resources/adult.data", header=None)
full_df.columns = ['age', 'workclass', 'fnlwgt', 'education', 'education-num', 'marital-status', 'occupation', 'relationship', 'race', 'sex', 'capital-gain', 'capital-loss', 'hours-per-week', 'native-country', 'income']
full_df.head()
```

```python
# Check datatypes and make sure no columns have nulls
full_df.info()
```

```python
# Check number of unique values in each column
unique_values = full_df.nunique()
print(unique_values)
```

```python
# Drop unnecessary columns
cleaned_df = full_df.drop(columns=['education'])
```

```python
# Check dataframe
cleaned_df.head()
```

```python
# Replace target column values with 0s and 1s
uniqueincome = cleaned_df['income'].unique()
print(uniqueincome)
```

```python
# Replace values in target column with 0 and 1
cleaned_df['income'] = cleaned_df['income'].map({' >50K': 1, ' <=50K': 0})
cleaned_df.head()
```

```python
# Export cleaned dataframe
cleaned_df.to_csv("../Resources/cleaned_df.csv")
```
